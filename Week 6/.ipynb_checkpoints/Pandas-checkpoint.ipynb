# Essential Functionality


### Reindexing
ðŸŽ¯An important method on pandas objects is reindex, which means to create a new object with the data confirmed to a new index

```python
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index = ['d', 'b','a', 'c'])
```

```python
obj
```

```python
# Calling reindex on this Series rearranges the data according to the new index, introducing missing values if any values were not already present
obj2 = obj.reindex(['a','b','c','d','e'])
```

```python
obj2
```

```python
# We can use the method parameter to fill the missing values automatically
obj3 = pd.Series(['blue','purple','yellow'], index = [0,2,4])
```

```python
obj3
```

```python
# method parameter can only be used when the index is ordered - meaning the index must have a clear order so Pandas knows what 'forward/backward' means
obj3.reindex(range(6), method = 'ffill')
```

```python
obj3.reindex(range(5), method = 'bfill')
```

```python
# With DataFrame, reindex can alter the (row) index, columns, or both. When passed only a sequence, it reindexes the rows in the result
frame = pd.DataFrame(np.arange(9).reshape(3,3), index = ['a','c','d'], columns = ['Ohio','Texas','California'])
```

```python
frame
```

```python
frame2 = frame.reindex(['a','b','c','d'])
```

```python
frame2
```

```python
# The columns can be reindexed with the columns keyword
states = ['Texas', 'Utah', 'California']
```

```python
frame.reindex(columns = states)
```

### Arithmetic Methods with Fill Values

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3,4)), columns = list('abcd'))
```

```python
df2 = pd.DataFrame(np.arange(20.).reshape((4,5)), columns = list('abcde'))
```

```python
df1
```

```python
df2.loc[1,'b'] = np.nan
```

```python
df2
```

```python
# Adding these together results in NA values in the locations that do not overlap
df1 + df2
```

```python
# Using the add() method on df1, we passed df2 and an argument to fill_value
df1.add(df2, fill_value = 0)
```

```python
df1.sub(df2, fill_value = 0)
```

```python
df1.mul(df2, fill_value = 0)
```

```python
df1.pow(df2, fill_value = 0)
```

```python
# Relatedly, when reindexing a Series or DataFrame, you can also specify a different fill value
df1.reindex(columns = df2.columns, fill_value = 0)
```

### Sorting and Ranking

```python
obj
```

```python
# Returns a new, sorted object
obj.sort_index()
```

```python
obj
```

```python
# With a DataFrame, we can sort by index on either axis
frame = pd.DataFrame(np.arange(8).reshape((2,4)), index = ['three','one'], columns = list('dabc'))
```

```python
# By default axis = 0 (vertically)
frame.sort_index()
```

```python
# Along columns
frame.sort_index(axis = 'columns')
```

```python
# Descending order
frame.sort_index(axis = 1, ascending = False)
```

```python
# To sort a Series by its valus, use its sort_values method
obj = pd.Series([4,7,-3,2])
```

```python
# Any missing values are sorted to the end of the Series by default
obj = pd.Series([4, np.nan, 7, np.nan])
```

```python
obj.sort_values()
```

```python
# When sorting a DataFrame, we can use the data in one or more columns as the sort keys. To do so, pass one or more column names to the 'by' option of sort_values
frame = pd.DataFrame({'b':[4,7,-3,2],'a':[0,1,0,1]})
```

```python
frame
```

```python
frame.sort_values(by = 'b')
```

```python
# Ranks can aslo be assigned according to the order in which they're observed in the data
obj.rank(method = 'first')
```

```python
# If the file contains a single-character delimiter, we can use Python's built-in csv muodule instead. To do so, we have to pass any open file or file like object to csv_reader()
import csv
with open('ex9.csv', 'r') as file:
    # Each row becomes a list of values with , separator
    reader = csv.reader(file)

    for line in reader:
        print(line)
```

```python
# We can do wrangling the data according to us
# Step 1: Read the file into a list of lines
with open('ex9.csv') as f:
    lines = list(csv.reader(f))
```

```python
# Step 2: Splitthe lines into header line and the data lines
header, values = lines[0], lines[1:]
```

```python
header
```

```python
values
```

```python
# Step 3: Create a dictionary of data columns using dictionary comprehension and the zip(*values), which transpose rows to columns
data_dict = {h:v for h,v in zip(header, zip(*values))}
```

```python
data_dict
```

```python
# There are many flavors of CSV files. So, to define a new format with a different delimiter, we define a simple subclass of csv.Dialect
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL

with open('ex9.csv') as f:
    reader = csv.reader(f, dialect = my_dialect)

    for line in reader:
        print(line)
```

```python
# In one line
with open('ex9.csv') as f:
    reader = csv.reader(f, delimiter = '|')

    for line in reader:
        print(line)
```

```python
# Write delimited file manually
with open('manual.csv', 'w') as f:
    writer = csv.writer(f, dialect = my_dialect)
    writer. writerow(('one','two','three'))
    writer.writerow((1,2,3))
    writer.writerow((4,5,6))
    writer.writerow((7,8,9))
```

### Discretization and Binning
ðŸ§±Splitting continuous data (like ages, income, scores, etc.) into intervals (bins) and assigning each value to a category (label)

```python
ages = [20,22,25,27,21,23,37,31,61,45,41,32]
```

```python
# Let's divide these into bins of 18 to 25,26 to 35,36 to 60 and finally 61 to older
# to do so, we have to use cut, a function in Pandas
bins = [18,25,35,60,100]
```

```python
cats = pd.cut(ages, bins)
```

```python
cats
```

```python
cats.codes #Retrns in which category the value felt
```

```python
cats.categories # Returns categories
```

```python
# Counts how many times each category/bin occurs
cats.value_counts()
```

```python
# We can change which side is closed by passing right = false
pd.cut(ages, [18,26,36,61,100], right = False)
```

```python
# We can also pass our own bin names by passing a list or array to the labels option
pd.cut(ages, bins, labels = ['Youth','YoungAdult','MiddleAged','Senior'])
```

```python
data = np.random.randn(20)
```

```python
data
```

```python
# If we specify an integer, Pandas automatically compute equal-length bins based on the minimum and maximum values in the data
pd.cut(data, 4, precision = 2) # precision = 2 option limits the decimal precision to two digits
```

```python
# A closely related function, qcut, bins the data based on sample quantiles, obtain roughply equal-size bins
data = np.random.randn(1000)
```

```python
new_data = pd.qcut(data, 4, labels = ['Q1','Q2','Q3','Q4']) # 4 equal quartiles
```

```python
new_data
```

```python
new_data.value_counts()
```

```python
# Similar to cut, we can pass our own quartiles (numbers between 0 to 1, inclusive)
pd.qcut(data, [0,0.1,0.5,0.9,1.])
```
