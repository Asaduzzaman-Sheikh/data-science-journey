## 7. Drop a column "city" and create a new column "age_group":
- "young" if age < 25
- "adult" otherwise

```python
df
```

```python
def get_age_group(age):
    if age < 20:
        return 'Teenager'
    elif age < 22:
        return 'Young Adult'
    else:
        return 'Adult'
```

```python
df['Age_Group'] = df['Age'].apply(get_age_group)
```
