# üé®Seaborn

A **Python library for making beautiful and advanced statistical graphs** with very simple code  

Think of it as:  
üëâMatplotlib = simple engine  
üëâSeaborn = beautiful, polished car built on top of the engine  

Most Seaborn functions use a DataFrame as input  
`sns.scatterplot(data = df, x = 'Age', y = 'Income')`
This is called **"tidy data" format**
- Each column = variable
- Each row = one observation

**Axes-level plots**  
There are plots that draw on a **single graph** (one Axes object)
- `sns.scatterplot()`
- `sns.lineplot()`
- `sns.barplot()`
- `sns.histplot()`
- `sns.boxplot()`
Think of these as **single-plot functions**

**Figure-level plots**  
These create a **full figure**, sometimes with multiple subplots automatically  
- `sns.relplot()`
- `sns.catplot()`
- `sns.displot()`
Think of these as functions that can create **many plots at once** or **multiple layouts**


**Categorical data**  
Data that is not numeric -- like labels or categories  
- Gender: "Male", "Female"
- City: "Dhaka", "Chittagong"
- Product: "A", "B", "C"

Seaborn has special plots for caegorical data:  
- `sns.catplot()`
- `sns.barplot()`
- `sns.countplot()`

**Continuous data**  
Numeric values that can take any number  
- `Age, Height, Income, Temperature`

Used plots like:  
- `sns.lineplot()`
- `sns.histplot()`
- `sns.kdeplot()`
- `sns.scatterplot()`

**Hue**  
Separates groups using **different colors**  
`sns.scatterplot(data = df, x = 'Age', y = 'Income', hue = 'Gender')`

**Style**  
Changes point style (shape) based on a category  
`sns.scatterplot(data = df, x = 'Age', y = 'Income', style = 'Gender')`  

**Size**  
Changes point size based on another variable  
`sns.scatterplot(data = df, x = 'Age', y = 'Income', size = 'Experience')`  

**Palatte**  
Color themes for Seaborn  
- `coolwarm`
- `viridis`
- `pastel`
- `deep`  
`sns.scatterplot(data = df, x = 'Age', y = 'Income', palatte = 'pastel')`

**KDE (Kernel Density Estimition)**  
A smooth version of a histgram  
`sns.kdeplot(data=df['Age'])`  

It gives a **smooth curve** showing the distribution  

**Pairplot**  
Creates **multiple plots at once** to compare all numeric columns  
`sns.pairplot(df)`  

Shows:  
- scatter plots for relationships
- histograms for distributions

Very useful for EDA  

**Heatmap**  
Displays values in a matrix using colors  
`sns.heatmap(df.corr(), annot=True)`  

Often used for correlation  

**regplot**  
Scatter plot **with a trend line**  
`sns.regplot(data=df, x='Age', y='Income')`  

Useful for seeing linear relationships  

**Themes/Styles**  
`sns.set_style("whitegrid")`  
`sns.set_theme(style="darkgrid")`  

Making it easier to get nice-looking plots quickly

### Lesson 1: The Huge Magic (`Color by Category`)
The single biggest reason people switch to Seaborn is the argument `hue`  

In Matplotlib, if you want to plot "Men" in Blue and "Women" in Orange, you have to write two separate `ax.scatter()` commands and handle the colors yourself. In seaborn, you just set `hue = 'Gender'`  

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#1. Load a built-in dataset (Seaborn has free data inside it!)
# This returns a Pandas DataFrame

df_tips = sns.load_dataset('tips')
display(df_tips.head())

#2. The Plot
# data = df_tips: Tell Seaborn where to look
# x, y          : The column names (Strings)
# hue           : The column to use for color
sns.scatterplot(data = df_tips, x = 'total_bill', y = 'tip', hue = 'sex')
plt.title("Tips vs Bill (Colored by Gender)")
plt.show()
```

### Lesson 2: The Heatmap(`sns.heatmap`)
Now let's move to the second superpower: **Correlations**. Scatter plots are great for 2-3 variables. But what if you have 10 variables? You can't draw 100 scatter plots  

You use a **Heatmap**. It colors the "Correlation Matrix" so you can spot patterns instantly  

```python
import pandas as pd

# The data
df = pd.DataFrame({
    'Speed': [100, 120, 140, 160, 180],
    'Weight': [1000, 1200, 1400, 1600, 1800], # Heavier cars are faster?
    'Efficiency': [30, 28, 25, 20, 15],       # Faster cars use more gas?
    'Price': [20, 25, 30, 45, 60]
})

# Calculate Correlation (The Math)
# This creates a table where 1.0 is perfect match and -1.0 is opposite
corr_matrix = df.corr()

print("--- The Correlation Matrix ---")
print(corr_matrix)

# Plot the Heatmap
plt.figure(figsize = (6,5))
# annot = Ture writes the numbers inside the boxes
# cmap = 'coolwarm' makes positive Red and Negative Blue
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.title("Car Specs Correlation")

plt.show()
```

You are analyzing a Student Performance dataset.

### Lesson 3: The Pairplot (`sns.pairplot`)
Now, imagine you have a dataset with 5 columns. You don't want to manually create 10 different scatter plots to check every ralationship  

**The Pairplot** does it all for you. It creates a grid of charts:  
- **Diagonals**: histograms (Distribution of that single variable)
- **Off-diagonals**: scatter plots (Relationship between two variables)

**Command**: `sns.pairplot(df)`

```python
# Load the data 
df_iris = sns.load_dataset('iris')

# The Magic Command
# hue = 'species' colors the dots by the flower type
sns.pairplot(df_iris, hue = 'species')

# plt.tight_layout()
plt.show()
```

### Lesson 4: The categorical Suite (`sns.boxplot`)
**Scatter plots** use numbers for X and Y. **Box plots** use a **Category** for X (`e.g., "Day of Week"`) and **Number** for Y (`e.g., "Total Bill"`)  

It answres: "What is the distribution of bills on Friday vs. Saturday?"  

**Comand**: `sns.boxplot(x = 'cat', y = 'num', data = df)`  


```python
# Load the data 
df_tips = sns.load_dataset('tips')

# X = category (Day)
# Y = Number (Bill)
sns.boxplot(data = df_tips, x = 'day', y = 'total_bill')

plt.title("Bill Distribution by Day")
plt.show()
```
