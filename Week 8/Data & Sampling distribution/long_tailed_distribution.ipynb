```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.stats as stats 
import seaborn as sns 
```

```python
df = sns.load_dataset('diamonds')
```

```python
# Information about the dataset... 
df.info()
```

```python
# Statistical summary... 
df.describe()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Plot - Histogram
plt.figure(figsize = (10, 6))
sns.histplot(df['price'], kde = True)
plt.show()
```

```python
log_price = np.log(df['price'])

# Histogram to check if the sahpe is now a bell curve
plt.figure(figsize = (8, 6))
sns.histplot(log_price, kde = True, color = 'teal')
plt.show()
```

```python
plt.figure(figsize = (8, 6))
sns.histplot(df, x = np.log(df['price']), hue = 'cut', kde = True)
plt.show()
```
