# The Bootstrap Method

```python
np.random.seed(42)
# Draw a sample of size 30 customers
population = np.random.exponential(scale = 40, size = 10000) # we assume after taking the sample, we can't access the population anymore, so we will use the sample to estimate 

# Draw one sample of size 30
sample_size = 30
sample_population = np.random.choice(population, size = sample_size, replace = True)

# From the sample_population take 1000 samples of size 30 
bootstrap_samples = np.random.choice(sample_population, size = (1000, sample_size), replace = True)

# Calculate the mean of each sample
bootstrap_means = np.mean(bootstrap_samples, axis = 1)


# Plot two histograms: one for the sample population and one for the sampling distribution of the mean
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
fig.suptitle('Sample Population vs Sampling Distribution of the Mean', fontsize = 16)

sns.histplot(sample_population, bins = 30, kde = True, color = 'skyblue', ax = ax[0])
ax[0].set_title('Sample Population')
ax[0].set_xlabel('Customer Spending ($)')
ax[0].set_ylabel('Frequency')

sns.histplot(bootstrap_means, bins = 30, kde = True, color = 'salmon', ax = ax[1])
ax[1].set_title('Sampling Distribution of the Mean')
ax[1].set_xlabel('Mean Values')
ax[1].set_ylabel('Frequency')

plt.show()

# Statistical Calculations
sample_mean = np.mean(sample_population)
population_mean = np.mean(population)
sample_std = np.std(bootstrap_means, ddof = 1)


print("=" * 30 + "Statistical Calculations" + "=" * 30)
print(f"Sample Mean: {sample_mean:.3f}")
print(f"Population Mean: {population_mean:.3f}")
print(f"Sample Standard Deviation: {sample_std:.3f}")
```

# Bootsrtap Confidence Interval

```python
# Calculate the 95% confidence interval for the sample mean using Bootstrap method
lower = np.percentile(bootstrap_means, 2.5)
upper = np.percentile(bootstrap_means, 97.5)
print(f"95% Confidence Interval: ({lower:.3f}, {upper:.3f})")

# CI using the classical method
confidence_level = 0.95
z_score = 1.96 # Z-score for 95% confidence
margine_of_error = z_score * ((np.std(sample_population, ddof = 1)) / np.sqrt(sample_size))

confidence_interval_lower = sample_mean - margine_of_error
confidence_interval_upper = sample_mean + margine_of_error

print(f"95% Confidence Interval using Classical Method: ({confidence_interval_lower:.3f}, {confidence_interval_upper:.3f})")
```

```python
# Plotting two CI's
plt.figure(figsize = (10, 6))

sns.histplot(bootstrap_means, bins = 30, kde = True, color = 'salmon', label = 'Bootstrap Means')

# Add vertical lines for the confidence intervals
plt.axvline(x = lower, color = 'red', linestyle = '--', label = 'Bootstrap Lower CI')
plt.axvline(x = np.mean(bootstrap_means), color = 'blue', linestyle = '--', label = 'Bootstrap Mean')
plt.axvline(x = upper, color = 'green', linestyle = '--', label = 'Bootstrap Upper CI')
plt.title('Bootstrap Sampling Distribution of the Mean with Confidence Intervals')
plt.xlabel('Mean Values')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

# Normal Distribution - The Bell Shape

```python
np.random.seed(42)

# Mean and standard deviation
mu = 66
sigma = 3

# Generate 1000 random heights with mean = 66, and std = 3
heights = np.random.normal(loc = mu, scale = sigma, size = 1000)
# print(heights.shape)

# Visualize the shape of the distribution
plt.figure(figsize = (10, 8))

sns.histplot(heights, bins = 30, kde = True, color = 'teal')
plt.show()
```

```python
# Calculate 1sigma
_1sigma = (heights >= 63) & (heights <= 69)

percentage = (np.mean(_1sigma)) * 100

# Calculate 2sigma
_2sigma = (heights >= 60) & (heights <= 72)
percentage1 = (np.mean(_2sigma)) * 100

# Calculate 3sigma
_3sigma = (heights >= 57) & (heights <= 75)
percentage2 = (np.mean(_3sigma)) * 100

print("="*70)
print(f"1 Sigma = {percentage}")
print(f"2 Sigma = {percentage1:.1f}")
print(f"3 Sigma = {percentage2:.1f}")


```

# The QQ  Plot

```python
import scipy.stats as stats 
```

```python
np.random.seed(42)

measurements = np.random.normal(loc = 0, scale = 1, size = 100)
measurements

# Visualizing
plt.figure(figsize = (12, 10))

stats.probplot(measurements, dist = 'norm', plot = plt)
plt.show()
```

```python
np.random.seed(42)

bad_data = np.random.uniform(low = -1, high = 1, size =100)

# Visualize
plt.figure(figsize = (10, 8))
stats.probplot(bad_data, dist = 'norm', plot = plt)
plt.show()
```

### The Z Score

```python
mu = 75
std = 10
score = 95

z_score = (score - mu) / std
# z_score

# Findd the percentile (probability)
percentage = stats.norm.cdf(z_score)
percentage
```

```python
# Load the messy field notes
df = sns.load_dataset('penguins')

# Pick the first 5 rows
df.head()
```

```python
plt.figure(figsize = (10, 6))

# Clean the data (remove missing values)
cleaned_weights = df['body_mass_g'].dropna()

sns.histplot(cleaned_weights, kde = True)
plt.title('The Eye Test: Penguin Weights')

plt.show()
```

```python
plt.figure(figsize = (6, 6))
stats.probplot(cleaned_weights, dist = 'norm', plot = plt)
plt.title("Q-Q Plot: Penguin Weights")
plt.show()
```

```python
# The Detective's Plot
sns.histplot(data=df, x='body_mass_g', hue='species', kde=True)
plt.title("The Hidden Tribes: Penguin Weights by Species")
plt.show()
```

```python
gentoo_data = df[df['species'] == 'Gentoo']['body_mass_g'].dropna()
from sympy.physics.units import gee

fig, ax = plt.subplots(1, 2, figsize = (10, 4), layout = 'constrained')
sns.histplot(gentoo_data, kde = True, color = 'g', ax = ax[0])
ax[0].set_title("Gentoo Weights (Histogram)")

stats.probplot(gentoo_data, dist = 'norm', plot = ax[1])
ax[1].set_title("Gentoo Weights (Q-Q Plot)")

plt.show()
```

```python
# Calculate mu and sigma
mu = gentoo_data.mean()
sigma = gentoo_data.std()

# Z Score 
observed_data = 5800
z_score = (observed_data - mu) / sigma

# Calculate the probabilty of a penguin being HAVIER than 5800g
""" 
    When you use the Python  command stats.norm.cdf(z_score), it calculates the percentiles:
        * Percentile = The line of people BEHIND you
        * But our question --> the line of people AHEAD of you
        So, the formula:
                total probability = 1
                => P(event occurs) + P(event does not occur) = 1
                
"""

total_probability = 1
prob_of_lighter = stats.norm.cdf(z_score)
prob_of_havier = total_probability - prob_of_lighter


print(f"Probability of a penguin being HAVIER than 5800g = {(prob_of_havier * 100):.2f}")
```
