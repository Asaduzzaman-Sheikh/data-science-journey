```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
# Generate the population of 10,000 data points
population = np.random.exponential(scale = 15, size = 10000)

# sample_size, n = 50
n = 50

# Generate 1000 samples of size n
samples = np.random.choice(population, size = (1000, n), replace = True)

# Calculate the mean of each sample
samples_mean = np.mean(samples, axis = 1)

# Plot the population distribution
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
fig.suptitle('Population vs Sampling Distribution of the Mean', fontsize = 16)

sns.histplot(population, bins = 30, kde = True, color = 'skyblue', ax = ax[0])
ax[0].set_title('Population Distribution')

sns.histplot(samples_mean, bins = 30, kde = True, color = 'salmon', ax = ax[1])
ax[1].set_title('Sampling Distribution of the Mean')

plt.show()

```

```python
# Use the population from the first challenge
np.random.seed(42)
population = np.random.exponential(scale = 15, size = 10000) 

# True population mean
population_mean = np.mean(population)

# Size of the sample, n = 50
n = 50

# Generate 1 samples of size 50
samples = np.random.choice(population, size = n, replace = True)
sample_mean = np.mean(samples)
sample_standard_deviation = np.std(samples, ddof=1)


print(f"Standard Deviation of the Sampling Distribution of the Mean: {sample_standard_deviation}")

# Calculate the 95% confidence interval for the sample mean
confidence_level = 0.95
z_score = 1.96 # Z-score for 95% confidence

margin_of_error = z_score * (sample_standard_deviation / np.sqrt(n))
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Check if the population mean is within the confidence interval
is_within_interval = (population_mean >= confidence_interval_lower) and (population_mean <= confidence_interval_upper)

print("=" * 70)
print(f"The true population mean is: {population_mean:.3f}")
print(f"Lower Bound: {confidence_interval_lower:.3f}")
print(f"Upper Bound: {confidence_interval_upper:.3f}")
print(f"Is the population mean within the confidence interval? {is_within_interval}")
```

```python
samples = np.random.choice(population, size = (1000,n), replace = True)
samples_mean = np.mean(samples, axis = 1)
samples_std = np.std(samples_mean, ddof=1)
print(f"Standard Deviation of the Sampling Distribution of the Mean: {samples_std:.3f}")

margin_of_error = z_score * (samples_std / np.sqrt(n))
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Lower Bound: {confidence_interval_lower:.3f}")
print(f"Upper Bound: {confidence_interval_upper:.3f}")

# Check if the population mean is within the confidence interval
is_within_interval = (population_mean >= confidence_interval_lower) and (population_mean <= confidence_interval_upper)

print(f"Is the population mean within the confidence interval? {is_within_interval}")

# Calculate the coverage rate
coverage_count = 0
lowers = []
uppers = []
captured = []

for i in range(1000):
    samples = np.random.choice(population, size = n, replace = True)
    sample_mean = np.mean(samples)
    sample_std = np.std(samples, ddof = 1)

    margin_of_error = z_score * (sample_std / np.sqrt(n))
    confidence_interval_lower = sample_mean - margin_of_error
    confidence_interval_upper = sample_mean + margin_of_error

    if (population_mean >= confidence_interval_lower) and (population_mean <= confidence_interval_upper):
        coverage_count += 1
        captured.append(True)
    else:
        captured.append(False)
    lowers.append(confidence_interval_lower)
    uppers.append(confidence_interval_upper)

# Plot the confidence intervals and indicate which ones captured the population mean
plt.figure(figsize = (10, 12))

for i in range(100):
    color = 'green' if captured[i] else 'red'
    plt.hlines(y = i, xmin = lowers[i], xmax = uppers[i], color = color, lw = 2)

plt.axvline(x = population_mean, color = 'blue', linestyle = '--', label = 'True Mean')
plt.xlabel('Tip Amount ($)')
plt.ylabel('Interval Number')
plt.title('First 100 Confidence Intervals')
plt.legend()
plt.show()


coverage_rate = coverage_count / 1000
print(f"Coverage Rate: {coverage_rate:.3f}")

captured_df = pd.DataFrame({
    'Lower Bound': lowers,
    'Upper Bound': uppers,
    'Captured': captured
})
print("=" * 30 + "Captured DataFrame" + "=" * 30)
captured_df
```

# Central Limit Theorem -- Code-First Exploration

## Basic Challenges
1. **Sample Means from a Skewed Distribution** -- Generate an exponential population, draw 1,000 samples of size 50, plot population vs sampling distribution of the mean.
2. **Effect of Sample Size** -- Compare sampling distributions for n = 2, 5, 30, 100 in a 2x2 grid.
3. **Quantify the Spread** -- Compute simulated vs theoretical standard error (` / n`) for various sample sizes.
4. **Visualize the Convergence** -- Line chart showing simulated and theoretical std overlap.
5. **The Final Test** -- Prove CLT works for any population shape (uniform, bimodal, exponential).

## Intermediate Challenges
6. **Build a Confidence Interval from Scratch** -- Compute a 95% CI from a single sample of 50.
7. **Does 95% Really Mean 95%?** -- Repeat 1,000 times, measure coverage rate, plot first 100 intervals.
8. **Effect of Confidence Level** -- Compare coverage rate and interval width across 80%, 90%, 95%, 99%.
9. **Effect of Sample Size on Interval Width** -- Fix confidence at 95%, vary n, show width decreases.
