```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(42)
# Generate the population of 10,000 data points
population = np.random.exponential(scale = 15, size = 10000)

# sample_size, n = 50
n = 50

# Generate 1000 samples of size n
samples = np.random.choice(population, size = (1000, n), replace = True)

# Calculate the mean of each sample
samples_mean = np.mean(samples, axis = 1)

# Plot the population distribution
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
fig.suptitle('Population vs Sampling Distribution of the Mean', fontsize = 16)

sns.histplot(population, bins = 30, kde = True, color = 'skyblue', ax = ax[0])
ax[0].set_title('Population Distribution')

sns.histplot(samples_mean, bins = 30, kde = True, color = 'salmon', ax = ax[1])
ax[1].set_title('Sampling Distribution of the Mean')

plt.show()

```

```python
# Sample sizes to compare
n1 = 2
n2 = 5
n3 = 30
n4 = 100

np.random.seed(42)

# Generate samples and calculate means for each sample size
samples_n1 = np.random.choice(population, size = (1000, n1), replace = True)
samples_mean_n1 = np.mean(samples_n1, axis = 1)

samples_n2 = np.random.choice(population, size = (1000, n2), replace = True)
samples_mean_n2 = np.mean(samples_n2, axis = 1)

samples_n3 = np.random.choice(population, size = (1000, n3), replace = True)
samples_mean_n3 = np.mean(samples_n3, axis = 1)

samples_n4 = np.random.choice(population, size = (1000, n4), replace = True)
samples_mean_n4 = np.mean(samples_n4, axis = 1)

# Plot the sampling distributions for different sample sizes
fig, ax = plt.subplots(2, 2, figsize = (12, 10))
fig.suptitle('Sampling Distribution of the Mean for Different Sample Sizes', fontsize = 16)

sns.histplot(samples_mean_n1, bins = 30, kde = True, color = 'skyblue', ax = ax[0, 0])
ax[0, 0].set_title(f'Sample Size = {n1}')

sns.histplot(samples_mean_n2, bins = 30, kde = True, color = 'salmon', ax = ax[0, 1])
ax[0, 1].set_title(f'Sample Size = {n2}')

sns.histplot(samples_mean_n3, bins = 30, kde = True, color = 'lightgreen', ax = ax[1, 0])
ax[1, 0].set_title(f'Sample Size = {n3}')

sns.histplot(samples_mean_n4, bins = 30, kde = True, color = 'lightcoral', ax = ax[1, 1])
ax[1, 1].set_title(f'Sample Size = {n4}')

plt.tight_layout(rect = [0, 0.03, 1, 0.95])
plt.show()
```

```python
# Calculate and print the means and standard deviations for different sample sizes for 1000 times
n = np.array([2, 5, 10, 30, 50, 100, 500])
times = 1000
means = [np.mean(np.random.choice(population, size = (times, size), replace = True), axis=1) for size in n]
stds = [np.std(m) for m in means]

# Population mean and standard deviation
population_mean = np.mean(population)
population_std = np.std(population)

# Theoretical standard deviation of the sampling distribution of the mean
theoretical_stds = population_std / np.sqrt(n)
```

```python
# Dataframe to display results
df = pd.DataFrame({
    'n': n,
    'simulated_std': stds,
    'theoretical_stds': theoretical_stds
})



# Plotting the simulated and theoretical standard deviations
plt.figure(figsize = (10, 6))
plt.plot(n, stds, marker = 'o', label = 'Simulated Std Dev', color = 'blue')
plt.plot(n, theoretical_stds, marker = 'o', label = 'Theoretical Std Dev', color = 'orange')
plt.xscale('log')
plt.xlabel('Sample Size (n)')
plt.ylabel('Standard Deviation')
plt.title('Simulated vs Theoretical Standard Deviation of the Sampling Distribution of the Mean')
plt.legend()
plt.grid()
plt.show()



# Plotting all the different sample sizes together
fig, axes = plt.subplots(2, 4, figsize = (20, 10))
fig.suptitle('Sampling Distributions of the Mean for Different Sample Sizes', fontsize = 16)

sns.histplot(means[0], bins = 30, kde = True, color = 'skyblue', ax = axes[0, 0])
axes[0, 0].set_title(f'Sample Size = {n[0]}')

sns.histplot(means[1], bins = 30, kde = True, color = 'salmon', ax = axes[0, 1])
axes[0, 1].set_title(f'Sample Size = {n[1]}')

sns.histplot(means[2], bins = 30, kde = True, color = 'lightgreen', ax = axes[0, 2])
axes[0, 2].set_title(f'Sample Size = {n[2]}')

sns.histplot(means[3], bins = 30, kde = True, color = 'lightcoral', ax = axes[0, 3])
axes[0, 3].set_title(f'Sample Size = {n[3]}')

sns.histplot(means[4], bins = 30, kde = True, color = 'lightblue', ax = axes[1, 0])
axes[1, 0].set_title(f'Sample Size = {n[4]}')

sns.histplot(means[5], bins = 30, kde = True, color = 'lightpink', ax = axes[1, 1])
axes[1, 1].set_title(f'Sample Size = {n[5]}')

sns.histplot(means[6], bins = 30, kde = True, color = 'teal', ax = axes[1, 2])
axes[1, 2].set_title(f'Sample Size = {n[6]}')
axes[1, 3].axis('off')  # Hide the empty subplot

plt.tight_layout(rect = [0, 0.03, 1, 0.95])
plt.show()

print("=" * 30 + "DataFrame" + "=" * 30)
df
```

```python
# Generate populations with different distributions
uniform_population = np.random.uniform(low = 0, high = 100, size = 10000)
bimodal_population = np.concatenate([np.random.normal(loc = 20, scale = 5, size = 5000), np.random.normal(loc = 80, scale = 5, size = 5000)])

# Plot the populations
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
fig.suptitle('Different Population Distributions', fontsize = 16)

sns.histplot(uniform_population, bins = 30, kde = True, color = 'skyblue', ax = ax[0])
ax[0].set_title('Uniform Distribution')

sns.histplot(bimodal_population, bins = 30, kde = True, color = 'salmon', ax = ax[1])
ax[1].set_title('Bimodal Distribution')

plt.tight_layout()
plt.show()
```

```python
# Sample size, n = 50
n = 50

# Generate samples and calculate means 
samples_bimodal = np.random.choice(bimodal_population, size = (1000, n), replace = True)
samples_mean_bimodal = np.mean(samples_bimodal, axis = 1)

# Population distribution and sampling distribution of the mean for bimodal population
fig, ax = plt.subplots(3, 2, figsize = (12, 18))
fig.suptitle('Visualization of Population and Sampling Distributions', fontsize = 16)

sns.histplot(bimodal_population, bins = 30, kde = True, color = 'teal', ax = ax[0, 0])
ax[0, 0].set_title('Bimodal Population')

sns.histplot(samples_mean_bimodal, bins = 30 , kde = True, color = 'lightcoral', ax = ax[0, 1])
ax[0, 1].set_title('Sampling Distribution of the Mean')

# Uniform population
samples_uniform = np.random.choice(uniform_population, size = (1000, n), replace = True)
samples_mean_uniform = np.mean(samples_uniform, axis = 1)

sns.histplot(uniform_population, bins = 30, kde = True, color = 'skyblue', ax = ax[1, 0])
ax[1, 0].set_title('Uniform Population')

sns.histplot(samples_mean_uniform, bins = 30, kde = True, color = 'salmon', ax = ax[1, 1])
ax[1, 1].set_title('Sampling Distribution of the Mean')

# Exponential population (original population)
samples_exponential = np.random.choice(population, size = (1000, n), replace = True)
samples_mean_exponential = np.mean(samples_exponential, axis = 1)

sns.histplot(population, bins = 30, kde = True, color = 'lightgreen', ax = ax[2, 0])
ax[2, 0].set_title('Exponential Population')

sns.histplot(samples_mean_exponential, bins = 30, kde = True, color = 'orange', ax = ax[2, 1])
ax[2, 1].set_title('Sampling Distribution of the Mean')

plt.tight_layout(rect = [0, 0.03, 1, 0.95])
plt.show()
```
