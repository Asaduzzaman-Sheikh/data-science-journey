```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Poisson & Related Distributions

```python
# Server Crash Analysis

"""
    Your server crashes on average 2 times per week.
    
    Question 1: What is the probability of exactly 0 crashes in a week?
    
    Question 2: What is the probability of exatcly 2 crashes in a week?

    Question 3: What is the probability of more than 4 crashes in a week?
"""

average_rate = 2
p_0 = stats.poisson.pmf(0, average_rate)
p_2 = stats.poisson.pmf(2, average_rate)
p_more_than_4 = 1 - stats.poisson.cdf(4, average_rate)

print("=" * 60)
print(f"P(0 crashes) = {p_0:.3f}")
print(f"P(2 crashes) = {p_2:.3f}")
print(f"P(more than 4 crashes) = {p_more_than_4:.3f}")
print("=" * 60)
```

# Scaling $lambda$

```python
# What about a whole month (4 weeks)?

# New_average_rate = lambda_original * (new_interval / original_interval)

new_average_rate = 2 * 4
p_more_than_12 = 1 - stats.poisson.cdf(12, new_average_rate)

print("=" * 60)
print(f"P(more than 12 crashes in a month) = {p_more_than_12:.3f}")
print("=" * 60)
```

# Exponential Distribution

```python
p_crash_within_2_days = stats.expon.cdf(2/7, scale=1/2)
p_no_crash_for_1_week = 1 - stats.expon.cdf(1, scale=1/2)

print("=" * 60)
print(f"P(crash within 2 days) = {p_crash_within_2_days:.3f}")
print(f"P(no crash for 1 week) = {p_no_crash_for_1_week:.3f}")
print("=" * 60)
```

# Weibull Distribution

```python
# Scenario: Light Bulb Lifetime
# A light bulb has an average life of 1,000 hours (λ = 1000). The failure rate increases with age (k = 2).

# Calculate P(bulb lasts less than 500 hours) using both distributions

pexp_bulb_lasts_less_than_500 = stats.expon.cdf(500, scale = 1000)
pwbl_bulb_lasts_less_than_500 = stats.weibull_min.cdf(500, c = 2, scale =  1000)

print(f"Exponential probability {pexp_bulb_lasts_less_than_500}")
print(f"Weibull probability {pwbl_bulb_lasts_less_than_500}")
print()
```
