```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Creating the dataset
data = {
    "Age": [22, 38, 26, 35, 35, 54, 2, 27, 14, 4],
    "Fare": [7.25, 71.28, 7.92, 53.10, 8.05, 51.86, 21.07, 11.13, 30.07, 16.70],
    "Pclass": [3, 1, 3, 1, 3, 1, 3, 3, 2, 3],
    "Survived": [0, 1, 1, 1, 0, 0, 0, 1, 1, 1]
}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
# Question 1: Age vs Fare
spearman_r, spearman_p = stats.spearmanr(df['Age'], df['Fare'])

print("="*60)
print(f"Spearman Correlation Coefficient r= {spearman_r:.3f}")
print(f"Spearman Correlation Coefficient p-value = {spearman_p:.3f}")
```

```python
# Question 2: Fare vs Passenger Class
kendall_r, kendall_p = stats.kendalltau(df['Fare'], df['Pclass'])

print("="*60)
print(f"Kendall Correlation Coefficient r= {kendall_r:.3f}")
print(f"Kendall Correlation Coefficient p-value = {kendall_p:.3f}")
```

```python
# Question 3: Age vs Survival

point_biserial_r, point_biserial_p = stats.pointbiserialr(df['Survived'], df['Age'])

print("="*60)
print(f"Point-Biserial Correlation Coefficient r= {point_biserial_r:.3f}")
print(f"Point-Biserial Correlation Coefficient p-value = {point_biserial_p:.3f}")
```
