```python
import numpy as np 
import scipy.stats as stats 
```

```python
observed = np.array([
    [50, 950],
    [65, 935]
])


chi_squared, p_value, dof, expected = stats.chi2_contingency(observed)
print("="* 60)
print(f"Chi squared : {chi_squared:.3f}" )
print(f"P value : {p_value:.3f}" )
```

# The Permutation Test

```python
drug =  [2, 3, 1, 4, 2]
placebo = [5, 6, 4, 7, 5]

observed_diff = np.mean(placebo) - np.mean(drug)
```

```python
# step 1: combine all 10 scores
all_scores = np.array(drug + placebo)

# step 2: loop 10,000 times 
times = 10000
count = 0
for i in range(times):
    np.random.shuffle(all_scores)
    perm_drug = all_scores[:5]
    perm_placebo = all_scores[5:]

    # step 3: difference
    perm_diff = np.mean(perm_drug) - np.mean(perm_placebo)

    # step 5: ask
    if perm_diff >= observed_diff:
        count += 1

p_value = count / times
print(f"Permutation p-value: {p_value}")
```

```python
t_stat, t_pvalue = stats.ttest_ind(drug, placebo)
print(f"t-test p-value: {t_pvalue:.3f}")
```

# Fisher's Exact Test

```python
# Treatment: 5 cured, 1 not cured
# Control: 1 cured, 5 not cured
table = np.array([[5, 1],
                  [1, 5]])

# Chi-Squared
chi2, chi_p = stats.chi2_contingency(table)[:2]

# Fisher's Exact
odds_ratio, fisher_p = stats.fisher_exact(table)

print("Chi‑squared test:")
print("chi2 =", chi2)
print("p =", chi_p)

print("\nFisher's exact test:")
print("odds ratio =", odds_ratio)
print("p =", fisher_p)
```
