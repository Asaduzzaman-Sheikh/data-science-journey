# Importing Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

## Set The Theme (Default)

```python
sns.set_theme()
```

## Exploring Seaborn Plots
* The main idea of Seaborn is that it provides high-level commands to ceate a variety of plot types useful for statistical data exploration, and even sxome statistical model fitting

### Histogram, KDE (kernal Density estimation), and densities

```python
# Load data 
df = sns.load_dataset('penguins')

# Create the histogram
fig, ax = plt.subplots(figsize = (8,5), layout = 'constrained', dpi = 300)

sns.histplot(data = df, x = 'flipper_length_mm', binwidth = 5, color = 'skyblue', ax = ax)
ax.set_title('Penguin Flipper Lengths (Histogrma)')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,5))

sns.histplot(data = df, x = 'flipper_length_mm', kde = True, color = 'purple')
ax.set_title('Histogram with KDE Overlay')

plt.show()
```

## Comparing Groups (The Power of `hue`)
* If you try to plot two histograms on top of each other in pure Matplotlib, it gets messy
* Seaborn handles transparency and stacking automatically

```python
fig, ax = plt.subplots(figsize = (10,6), dpi = 300, layout = 'constrained')

sns.histplot(data = df, x = 'flipper_length_mm', hue = 'species',
            element = 'step')
ax.set_title('Flipper Length by Species')

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,6), dpi = 300, layout = 'constrained')

sns.histplot(data = df, x = 'flipper_length_mm', hue = 'species',
            multiple = 'stack')
ax.set_title('Flipper Length by Species')

plt.show()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], density = True, alpha = 0.5)
plt.title('Histogram for Visualizing Distribution')
plt.show()
```

```python
"""
Rather than just providing a histogram as a visual output, we can get a smooth estimate
of the distribution using kernel density estimation
"""
sns.kdeplot(data = data, fill = True)
plt.show()
```

## Categorical Plots
* When one variable is categorical (like "Species", or "day of Week") and other is numerical (like "Price", or "Weight"), you need **Categorical Plots**
  - **Boxplot**: The statistical summary
  - **Violin Plot**: The density shape
  - **Swarm Plot**: The raw data points

### The Boxplot (`sns.boxplot`)
* Summarizes data into five numbers: *Minimum, First Quartile(25%), Median(50%), Third Quartile(75%),and MAximum*

```python
fig,ax = plt.subplots(figsize = (8,6))

sns.boxplot(data = df, x = 'species', y = 'body_mass_g', hue = 'species', palette = 'Set2', ax = ax)
ax.set_title('Body Mass Distribution by Species')

plt.show()
```

### The Violin Plot (`sns.violinplot`)
* Essentially a **Boxplot + KDE**
* Mirrors the density curve to create a symmetrical shape

```python
fig,ax = plt.subplots(figsize = (10,6))

# split=True: Instead of two separate violins, it draws one split down the middle!
# This is incredibly space-efficient.
sns.violinplot(data = df, x ='species', y ='body_mass_g', hue = 'sex', split = True, inner = 'quart', palette = 'muted', ax = ax)

ax.set_title('Violin Plot: Mass by Species & Sex')

plt.show()
```

### The Swarm Plot (`sns.swarmplot`)
* Draws every single data point but adjust their positions so they don't overlap

**Pro Tip**: Often, the best visualization is combining a *Boxplot* with a *Swarmplot*

```python
fig,ax = plt.subplots(figsize = (8,6))



# 1. Draw the Boxplot (Make it faint so points stand out)
sns.boxplot(data=df, x="species", y="body_mass_g", color="white", ax = ax)

# 2. Draw the Swarm Plot on top
sns.swarmplot(data=df, x="species", y="body_mass_g", color="black", size=4, alpha=0.7)

ax.set_title("Boxplot + Swarmplot (The Complete Picture)")
plt.show()
```

### Module 4: Matrix Plots (Heatmap)
* A **Heatmap** essentially a grid where individual values are represented by colors
  - **X-axis**: Category A(`e.g., Year`)
  - **Y-axis**: Category B(`e.g., Month`)
  - **Color**: The values (`e.g., Number of passengers`)
* The most common use case in Data Science is the **Correlation Matrix**, which tells you how strongly different variables are related to each other



```python
# Calculate Correlation 
# This creates a square grid (Matrix)
corr_matrix = df.corr(numeric_only = True)

# Draw the Heatmap
fig,ax = plt.subplots(figsize = (8,6), dpi = 300, layout = 'constrained')

# data: The matrix we just calculated
# annot=True: Writes the actual number inside the box (Crucial for reading!)
# cmap='coolwarm': A standard color map (Blue=Negative, Red=Positive)
# vmin=-1, vmax=1: anchors the colors so White is exactly 0
sns.heatmap(data = corr_matrix, annot = True, cmap = 'coolwarm', vmax = 1, vmin = -1, ax = ax)

ax.set_title('Correlation Matrix of Penguin Features')

plt.show()
```

```python
# The Pivot Table heatmap (Time Series)
# Load the 'flights' dataset
flights = sns.load_dataset('flights')

# 1. PIVOT THE DATA
# We want: Month on Y-axis, Year on X-axis, Passengers as Value
flight_matrix = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')

fig,ax = plt.subplots(figsize = (10,6))

# linewidths=.5: Adds small white lines between squares to separate them
# fmt="d": Formats the annotation as 'decimal' (integers) instead of scientific notation
sns.heatmap(data = flight_matrix, cmap = 'YlGnBu', linewidth = 0.5, ax = ax)

ax.set_title("Flight Passengers (1949 - 1960)")
plt.show()
```

### Module 5: The 'All-in-One' Plots (Pairplot & Jointplot) 

#### The Pairplot (`sns.pairplot`)

```python
sns.pairplot(df, hue = 'species', palette = 'husl')

plt.show()
```

#### The Jointplot (`sns.jointplot`)

```python
# kind = 'reg': Adds a linear regression line (trend line) and a confidence interval
# height = 7: makes the plot square (7 X 7 inches)
sns.jointplot(data = df, x = 'flipper_length_mm', y = 'body_mass_g', kind = 'reg', color = 'm', height = 7)

plt.show()
```

### Module 6: Aesthetics & Customization

**The Background Style** (`set_style`)


```python
# Set the style
sns.set_style('whitegrid')
```

```python
fig,ax = plt.subplots(figsize = (8,5))
sns.boxplot(data = df, x = 'species', hue = 'species', y = 'body_mass_g', palette = 'pastel', ax = ax)

# 3. CLEAN UP BORDERS (The "Despine" Trick)
# If you use the 'white' or 'ticks' style, the top and right box lines can look ugly.
# This command removes them instantly.
sns.despine(left=True)

ax.set_title("Clean & Minimal Style")
plt.show()
```

**The Context** (`set_context`)


```python
fig,ax = plt.subplots(figsize = (8,5))
sns.set_context('talk')

sns.scatterplot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', ax = ax)

ax.set_title('Scaled for a talk (Big Fonts)')

plt.show()
```

**Color Palette** (`palette`)


```python
# Reset context
sns.set_context("notebook")

# DRAW WITH CUSTOM COLORS
plt.figure(figsize=(8, 5))

# We can name a palette string, OR pass a list of hex codes!
# palette="magma" is a beautiful purple-to-yellow gradient
sns.barplot(data=df, x="species", y="body_mass_g",hue = 'species',  palette="magma")
sns.despine()

plt.title("Using the 'Magma' Palette")
plt.show()
```

### Module 7: The faceted Plots (`relplot,catplot,distplot`)

```python
tips = sns.load_dataset('tips')

# 1. CREATE A FACETED SCATTER PLOT
# col="time": This splits the graph into two side-by-side plots (Lunch vs Dinner)
sns.relplot(
    data=tips, 
    x="total_bill", 
    y="tip", 
    hue="smoker",   # Color dots by smoker status
    col="time",     # Create a separate plot for each Time
    style="smoker", # Change dot shape for smokers
    height=5,       # Height of EACH subplot
    aspect=1        # Aspect ratio (width/height)
)

plt.show()
```

**The 2D Facet Grid (Rows AND Columns)**

```python
# 1. CREATE A FACETED HISTOGRAM
# row="sex": Male/Female are rows
# col="time": Lunch/Dinner are columns
sns.displot(
    data=tips, 
    x="tip", 
    col="time", 
    row="sex", 
    kde=True,       # Add the smooth density line
    height=3,       # Make them small so they fit on screen
    color="teal"
)

plt.show()
```
