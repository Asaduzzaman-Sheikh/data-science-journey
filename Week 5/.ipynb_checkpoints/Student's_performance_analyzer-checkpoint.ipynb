```python
import numpy as np

# Create a 10 x 3 array of random integers between 0 and 100
data = np.random.randint(0, 101, size = (10, 3))

# Assign columns
columns = ['Math', 'Physics', 'CS']

# Print original data
print('******************************************************')
print(columns)
print("Original Data:\n", data)
print('******************************************************')
print()


# Compute per student average
student_averages = np.mean(data, axis = 1)
print('******************************************************')
print("\nStudent Averages:\n", student_averages)
print('******************************************************')
print()

# Compute per subject average
subject_averages = np.mean(data, axis = 0)
print('******************************************************')
for name, marks in zip(columns, subject_averages):
    print(f"Average marks in {name}: {marks}")
print('******************************************************')
print()
# Student with highest average
top_student = np.max(student_averages)
# top_student_index = np.argmax(student_averages)

# Subject with highest average
top_marks_subject = np.max(subject_averages)
# top_subject_index = np.argmax(subject_averages)
print('******************************************************')
print("\nHighest Student Average:", top_student)
print("Highest Subject Average:", top_marks_subject)
print('******************************************************')
print()
# Normalize data to range [0, 1]
normalized_data = (data - np.min(data)) / (np.max(data) - np.min(data))

print('******************************************************')
for i in range(len(data)):
    print(f"Student {i + 1} - Original Marks: {data[i]}, Normalized Marks: {normalized_data[i]}")
print('******************************************************')
```
