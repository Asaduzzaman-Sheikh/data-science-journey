
6ï¸âƒ£ Stacking Arrays
-------------------
Horizontally and vertically stack two random 3Ã—3 arrays.

```python
arr1 = np.random.random((3,3))
arr2 = np.random.random((3,3))
print(arr1.shape)
print(arr2.shape)
print()

print(arr1)
print()

print(arr2)
print()

# Horizontally stack
stack_h = np.hstack((arr1,arr2))

# Vertically stack
stack_v = np.vstack((arr1,arr2))

print(stack_h)
print()

print(stack_v)
print()
```

8ï¸âƒ£ Normalization
-----------------
Given a random 1-D array, normalize it to the range 0â€“1.  
Basic Syntax: normalized = a + ( (data - np.min(data)) * (b - a) ) / (np.max(data) - np.min(data)); where a, b the ranges

```python
array = np.random.randint(1,100,10)
print(array)
print()

# Range a = 0 - b = 1
normalized = 0 + ((array - np.min(array)) * (1 - 0)) / (np.max(array) - np.min(array))
print(normalized)
```

9ï¸âƒ£ Random Sampling
-------------------
Generate a 5Ã—5 matrix with random integers 0â€“100.
Replace all numbers greater than 50 with 50.

```python
array = np.random.randint(0,100,25)
print(array.shape)

reshaped = np.reshape(array, (5,5))
print(reshaped.shape)
print()

print(array)
print(reshaped)
print()

mask = reshaped > 50
reshaped[mask] = 50
print(reshaped)
```
