<h1>Bivariate Analysis</h1>
<p>Bivariate data refers to a dataset where each observation is associated with two different variables. The goal of analyzing bivariate data is to understnd the relationship or association between tese two variables</p>
<ul>
    <li>Detects whether the relationship is positive, negative or non-existent</li>
    <li>Correlation measures the strength and direction of the relationship (range: -1 to +1)</li>
    <li>Visualizatin tools like scatter plots and regression lines show the pattern of change clearly</li>
    <li>Often used as a foundation before moving to multivariate models</li>
</ul>

<h1>Multivariate Analysis</h1>
<p>Multivariate data contains three or more variables for each observation. The objective is to how multiple variables interact or jointly affect outcomes. It's crucial in fields like predictive analytics, econometrics and data science, where relationships are seldom limited to two variables</p>
<ul>
    <li>Useful when an outcome depends on several influencing factors</li>
    <li>Techniques include multiple regression, PCA, MANOVA and clustering</li>
    <li>Reduces data complexity using dimensionality reduction (like PCA)</li>
    <li>Commonly visualized using heatmaps, pair plots, and 3D scatter plots for high-dimensional insight</li>
</ul>

```python
import pandas as pd
import  seaborn as sns
```

```python
tips = sns.load_dataset('tips')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

<h2>Scatter Plots (Numeric - Numeric)</h2>
<p>Scatter plots are one of the most fundamental tools for visualizing rlationships between two numerical variables</p>

```python
tips.sample(5)
```

```python
# bivariate analysis
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
# multivariate analysis
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', data = tips)
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size', data = tips)
```

<h2>Bar Plot (Numerical - Categorical)</h2>
<p>A bat plot uses tectangular bars to represents data categories, with bar length or height propotional to their values. It compares discrete categories, with one axis for categories and other for values</p>

```python
titanic.sample(5)
```

```python
sns.barplot(x = 'pclass', y = 'age', data = titanic)
```

```python
sns.barplot(x = 'pclass', y = 'fare', data = titanic)
```

```python
sns.barplot(x = 'pclass', y = 'fare', hue = 'sex', data = titanic)
```

<h2>Bar Plot (Numerical - Categorical)</h2>

```python
sns.boxplot(x = 'sex', y = 'age', data = titanic)
```

```python
sns.boxplot(x = 'sex', y = 'age', hue = 'survived', data = titanic)
```

<h2>Dist Plot (Numerical - Categorical)</h2>

```python
sns.distplot(titanic[titanic['survived'] == 0]['age'], hist = False)
sns.distplot(titanic[titanic['survived'] == 1]['age'], hist = False)
```

<h2>Heatmap (Categorical - Categorical)</h2>
<p>Heatmaps arae a powerful data visualization tool that allows you to represent data in a matrix format using colors. They are particularly useful for identifying patterns, correlations, and outliers in large datasets</p>

```python
sns.heatmap(pd.crosstab(titanic['pclass'], titanic['survived']))
print(pd.crosstab(titanic['pclass'], titanic['survived']))
```

```python
# survived by percentages
(titanic.groupby('pclass')['survived'].mean() * 100).plot(kind = 'bar')
print(titanic.groupby('pclass')['survived'].mean() * 100)
```

```python
titanic.groupby('sex')['survived'].mean() * 100
```

<h2>Cluster Map (Categorical - Categorical)</h2>
<p>Plots a hierarchical-clustered heat map of the given matrix dataset. It returns a clustered grid index</p>

```python
sns.clustermap(pd.crosstab(titanic['parch'], titanic['survived']))
```

<h2>Pair Plot</h2>
<p>Used for visualizing relationships between multiple variables in a dataset. By creating a grid of scatter plots it helps to identify gow different features interact with each other toidentify patterns, correlations and trends in data</p>

```python
iris.sample(5)
```

```python
sns.pairplot(iris, hue = 'species')
```

<h2>Line Plot (Numerical - Numerical)</h2>

```python
flights.sample(5)
```

```python
data = flights.groupby('year')['passengers'].sum().reset_index()
sns.lineplot(x = 'year', y = 'passengers', data = data)
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```
