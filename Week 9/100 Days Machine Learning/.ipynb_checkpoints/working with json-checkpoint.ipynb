# Topic 1 - Reading JSON with `pd.read_json()`

JSON data comes in many shapes, Pandas has a parameter called `orient` that tells it what structure to expect

```python
import pandas as pd
import json
```

```python
# Most common: a list of records (each dict = one row)
data = '[{"name": "Alice", "score": 92}, {"name": "Bob", "score": 85}]'
df = pd.read_json(data)
df
```

```python
# Reading from a file
df = pd.read_json('train.json', orient = 'records')
df
```

```python
# Reading from a URL directly
df = pd.read_json("https://jsonplaceholder.typicode.com/posts")
df
```

```python
df = pd.read_json("https://jsonplaceholder.typicode.com/comments")
df
```

```python
df = pd.read_json("https://jsonplaceholder.typicode.com/users")
df
```

**Always specify `orient` explicitly.** Pandas will guess if you omit it, but it gets wrong on edge cases. Always declare it

# Topic 2 - Flattening Nested JSON with `json_normalize()`

```python
from pandas import json_normalize 
data = [
    {
        "id": 1,
        "student": {"name": "Alice", "grade": "A"},
        "scores": [92, 88, 95]
    },
    {
        "id": 2,
        "student": {"name": "Bob", "grade": "B"},
        "scores": [75, 80, 78]
    }
]

# Basic normalize - flattens one level of nesting
df = json_normalize(data)
print(df.columns.tolist())
print("\nNotice student.name — Pandas uses dot notation to name the flattened columns. This is the default separator (you can change it).")
df
```

## Handling arrays inside records - `record_path`

* `record_path`*: defines the core data you want to turn into rows
* `meta`: adds accompanying contextual information for higher levels in the JSON to those rows (Parent-level fields to carry into each expanded row)
* `sep`: change column separator (default `.`, can use `_`)
* `errors`: `raise` (default) or `ignore` missing meta keys
* `max_level`: how many levels deep to flatten (great for controlling depth)

```python
# Explode the nested 'scores' array into its own rows
df = json_normalize(data, record_path = ['scores'], meta = ['id',['student', 'name']])
df
```

```python
# Change separator from dot to underscore
df = pd.json_normalize(data, sep = '_')
df
```

**`max_level` is your safety net**. On deeply nested JSON (4–5 levels), normalizing everything explodes your column count. Set `max_level=2` to flatten only what you need and keep deeper nesting as dicts in a column.

# Topic 3 - JSON Stored Inside a DataFrame Column

This is very common when your data comes from a database (e.g., PostgreSQL JSONB columns) where each row has a JSON string in one cell.

```python
import ast 
df = pd.DataFrame({
    "user_id": [1, 2, 3],
    "metadata": [
        '{"city": "Dhaka", "age": 22}',
        '{"city": "Chittagong", "age": 25}',
        '{"city": "Sylhet", "age": 30}'
    ]
})
df
```

```python
# Step 1: Parse the string column into actual dicts
df['metadata_parsed'] = df['metadata'].apply(json.loads)
df
```

```python
# Step 2: Expand those dicts into real columns
df = df.join(pd.json_normalize(df['metadata_parsed']))
df
```

`json.loads` vs `ast.literal_eval` — **know the difference**:  

 * `json.loads()`: use when the string is valid JSON (uses double quotes)ast.
 * `literal_eval()`: use when the string looks like a Python dict (single quotes)


```python
# json.loads works on: '{"key": "value"}'
# ast.literal_eval works on: "{'key': 'value'}"
# df['parsed'] = df['metadata'].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
# df
```

```python
# always guard against nulls before parsing
# df['parsed'] = df['metadata'].apply(lambda x: json.loads(x) if pd.notna(x) else {})
# df
```

```python
# The Bulletproof Solution
def safe_json_parse(x):
    if isinstance(x, str):
        return json.loads(x)
    elif pd.isna(x):
        return {}
    return x

# Apply the robust function
df['parsed'] = df['metadata'].apply(safe_json_parse)
df
```

<h1>Topic 5 - Writing DataFrames to JSON with <i>to_json()</i></h1> 

```python
df = pd.DataFrame({
    "name": ["Alice", "Bob"],
    "score": [92, 85]
})

# write to a file
df.to_json('output.json', orient = 'records', indent = 2)

# get JSON as a string
json_str = df.to_json(orient = 'records')
print(json_str)
```

<h1>Topic 6 - Edge Cases</h1>

```python
# missing keys in some records
data = [
    {"name": "Alice", "score": 92, "grade": "A"},
    {"name": "Bob", "score": 85},  # no "grade" key!
]

df = pd.DataFrame(data)

# Pandaas fills the missing key with NaN automatically - but only if all records are dicts
df
```

```python
# if some are None, you need to guard
df = pd.DataFrame([r for r in data if r is not None])
df
```
