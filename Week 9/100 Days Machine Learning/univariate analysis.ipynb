<h1>Univariate Analysis</h1>
<p>Univariate analysis is a type of data visualization where we visualize only a single variable at a time. Univariate analysis helps us to analyze the distribution of the variable present in the data so that we can perform further analysis</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Employee_dataset.csv')
df.sample(5)
```

<h3>Count Plot</h3>
<p>seaborn.countplot() is a founction used to display the counts of observations in categorical data. It shows the distribution of a single categorical variable or the relationship between two categorical variables by creating bar plot</p>

```python
sns.countplot(x = 'gender_full', data = df)
```

```python
df['gender_full'].value_counts().plot(kind = 'bar')
df['gender_full'].value_counts()
```

<h3>Pie Chart</h3>
<p>Helps us to visualize the percentage of the data belonging to each category</p>

```python
df['gender_full'].value_counts().plot(kind = 'pie', autopct = '%.1f%%')
df['gender_full'].value_counts()
```

```python
df['STATUS_YEAR'].value_counts().plot(kind = 'pie', autopct = '%.1f%%')
```

<h3>Histogram</h3>
<p>We'll be performing univariate analysis on Numerical variables using histogram function</p>

```python
sns.histplot(x = 'age', bins = 10, data = df)
```

<h3>Distplot</h3>

```python
sns.distplot(df['age'])
```

<h3>Boxplot</h3>

```python
sns.boxplot(df['age'])
```

```python
sns.boxplot(x = 'age', data = df)
```
